# Data Preprocessing: Normalization and Standardization

## Introduction

In machine learning, preprocessing is a crucial step to ensure that the data is suitable for modeling. Two common preprocessing techniques are **normalization** and **standardization**. Both methods adjust the scale of features in the dataset, which can significantly impact the performance of various machine learning algorithms.

## Concepts

### 1. Normalization (Min-Max Scaling)

**Definition:**
Normalization rescales the features of the data to fit within a specified range, usually [0, 1].

**Purpose:**
- Ensures that all features have the same scale, which is particularly important for algorithms that are sensitive to feature scales (e.g., neural networks).
- Helps to improve the convergence of optimization algorithms.

**Impact:**
- Changes the range of the data but preserves the distribution's shape.

### 2. Standardization (Z-Score Normalization)

**Definition:**
Standardization transforms features to have a mean of 0 and a standard deviation of 1.

**Purpose:**
- Centers the data around zero, which is beneficial for many machine learning algorithms that assume normally distributed data (e.g., linear regression, logistic regression).
- Scales the data so that each feature contributes equally to the model.

**Impact:**
- Centers the data but preserves the shape of the distribution.
- Useful when features have different units or scales.

## Code Explanation





### 1. Data Loading

```python
import pandas as pd

# Load data
df = pd.read_csv('https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv', header=None, usecols=[0,1,2])
df.columns = ["Class", "Alcohol", "Malic"]
df.head()
```

### 2. Normalization

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize data
scaling = MinMaxScaler()
df_normalized = df[['Alcohol', 'Malic']].copy()
df_normalized[['Alcohol', 'Malic']] = scaling.fit_transform(df[['Alcohol', 'Malic']])
```

### 3. Standardization

```python
from sklearn.preprocessing import StandardScaler

# Standardize data
standardization = StandardScaler()
df_standardized = df[['Alcohol', 'Malic']].copy()
df_standardized[['Alcohol', 'Malic']] = standardization.fit_transform(df[['Alcohol', 'Malic']])
```

### 4. Visualization

Histograms

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 5))

# Original data
plt.subplot(1, 3, 1)
plt.hist(df['Alcohol'], bins=30, alpha=0.7, label='Alcohol')
plt.hist(df['Malic'], bins=30, alpha=0.7, label='Malic')
plt.title('Original Data')
plt.legend()

# Normalized data
plt.subplot(1, 3, 2)
plt.hist(df_normalized['Alcohol'], bins=30, alpha=0.7, label='Alcohol')
plt.hist(df_normalized['Malic'], bins=30, alpha=0.7, label='Malic')
plt.title('Normalized Data')
plt.legend()

# Standardized data
plt.subplot(1, 3, 3)
plt.hist(df_standardized['Alcohol'], bins=30, alpha=0.7, label='Alcohol')
plt.hist(df_standardized['Malic'], bins=30, alpha=0.7, label='Malic')
plt.title('Standardized Data')
plt.legend()

plt.tight_layout()
plt.show()
```

Scatter Plots

```python
plt.figure(figsize=(18, 5))

# Original data scatter plot
plt.subplot(1, 3, 1)
plt.scatter(df['Alcohol'], df['Malic'], alpha=0.7)
plt.title('Original Data')

# Normalized data scatter plot
plt.subplot(1, 3, 2)
plt.scatter(df_normalized['Alcohol'], df_normalized['Malic'], alpha=0.7)
plt.title('Normalized Data')

# Standardized data scatter plot
plt.subplot(1, 3, 3)
plt.scatter(df_standardized['Alcohol'], df_standardized['Malic'], alpha=0.7)
plt.title('Standardized Data')

plt.tight_layout()
plt.show()
```

## Conclusion
Normalization and standardization are essential techniques in data preprocessing. They ensure that features are on a comparable scale, which can enhance the performance and stability of machine learning models. By applying these techniques and visualizing their effects, you can better understand how data transformations impact your dataset.
